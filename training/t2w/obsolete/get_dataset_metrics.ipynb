```python
import pandas as pd
```

## D1a

```python
df = pd.read_csv('D1a.tsv', sep='\t')
display(df.head())
print(f"Number of images: {len(df)}, from {len(set([name.split('_')[0] for name in df['file_name'].tolist()]))} distinct subjects")
print(f"Sex proportion: M = {len(df[df['sex']=='M'])/len(df):.2f}, F = {len(df[df['sex']=='F'])/len(df):.2f}")
print(f"Age mean: {df['age'].mean():.2f}, std: {df['age'].std():.2f}")
print(f"Neck flexion proportion: Up = {len(df[df['neck_flexion'] == 'U'])/len(df):.2f}, Normal = {len(df[df['neck_flexion'] == 'N'])/len(df):.2f}")
```
